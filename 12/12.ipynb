## Visualise day 12
Initally, initalise the data to be used

```python
import numpy as np

file = '12_input.txt'
with open(file) as f:
    data = f.read().splitlines()

heightmap = np.zeros((len(data), len(data[0])), int)
height_lists = np.zeros((len(data) * len(data[0]), 3))

iter = 0
                        
for row, line in enumerate(data):
    for col, value in enumerate(line):
        value = ord(value)
        if value == 83:
            # start, S
            value = 0
        elif value == 69:
            # end, E
            value = 25
        else:
            value -= 97
        heightmap[row, col] = value
        height_lists[iter] = [row, col, value]
        iter +=1
        
# print(heightmap)
# print(height_lists)
```

#### Now visualise the heightmap

```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go

# Define Data

x_max = heightmap.shape[0] - 1
y_max = heightmap.shape[1] - 1
```

```python
plt.imshow(heightmap)
plt.colorbar()
plt.show()
```

```python
x = height_lists[:, 0]
y = height_lists[:, 1]
z = height_lists[:, 2]

# add height map surface
fig = plt.figure(figsize=(25, 15))
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z)
```

```python
fig = plt.figure(figsize=(25, 15))
ax = fig.add_subplot(projection='3d')
ax.plot_trisurf(x, y, z, linewidth=0.2, antialiased=False, cmap='gnuplot')

# cmap options:
# ['flag', 'prism', 'ocean', 'gist_earth', 'terrain',
#                       'gist_stern', 'gnuplot', 'gnuplot2', 'CMRmap',
#                       'cubehelix', 'brg', 'gist_rainbow', 'rainbow', 'jet',
#                       'turbo', 'nipy_spectral', 'gist_ncar'])
```
